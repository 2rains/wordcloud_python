```python
# with open('allList.txt','r',encoding='utf-8') as d:
with open('allList.txt','r',encoding='utf-8') as text:
    text1 = text.read()
#     text1
```

```python
# NLTK 자연어분석
import nltk
from konlpy.tag import Okt
t = Okt()
```

```python
present_text = ''

for each_line in text1[:len(text1)]:
    present_text = text1 + each_line + '\n'

# new line으로 나눠서 구분
```

```python
# 형태소 분석
tokens_ko = t.morphs(present_text)
tokens_ko
```

```python
len(tokens_ko)
```

```python
# 형태소 -> 토큰 -> 실 토큰 수(중복 제외)
ko = nltk.Text(tokens_ko, name='a')
print(len(ko.tokens))
print(len(set(ko.tokens)))
```

```python
# 빈도수 확인
ko = nltk.Text(tokens_ko, name='a')
ko.vocab().most_common(100) # 빈도수 상위 100개
```

```python
# 불용어 처리
stop_words = ['.','가','요','답변','...','을','수','에','질문','제','를','이','도',
                      '좋','1','는','로','으로','2','것','은','다',',','니다','대','들',
                      '2017','들','데','..','의','때','겠','고','게','네요','한','일','할',
                      '10','?','하는','06','주','려고','인데','거','좀','는데','~','ㅎㅎ',
                      '하나','이상','20','뭐','까','있는','잘','습니다','다면','했','주려',
                      '지','있','못','후','중','줄','6','과','어떤','기본','!!',
                      '단어','선물해','라고','중요한','합','가요','....','보이','네','무지',
                      '****','!','(',')','개','저','인지','하는데','같아요','적','위',
                      '된','인','인거','하고','이라고','그','적','합니다','여러','없는데',
                      '괜찮은','해주세요','해주려고','입니다','안녕하세요','아니더라도',
                      '괜찮고','괜찮을지','몰라서요','그러는데','\n','\n \n','\n\n\n',' ','ㅠ','내공','(','에서','알려주세요','와','찾아주세요','인가요']
```

```python
# 불용어 반복 처리
tokens_ko = [each_word for each_word in tokens_ko
                        if each_word not in stop_words]
```

```python
ko = nltk.Text(tokens_ko, name='char')
ko.vocab().most_common(50) # 빈도수 높은 단어 상위 50개
```

```python
# 그래프
plt.figure(figsize=(15,6))
ko.plot(50)
plt.show()
```
